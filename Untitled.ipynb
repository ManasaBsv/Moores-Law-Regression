```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#Read the data
data = pd.read_csv('moore.txt', header = None)
```

```python
#See the data
data.head()
```

```python
#Data shape
data.shape
```

```python
# x-DATA and Y-DATA, reshape X into a 2D array, as tf expects a 2D input array while training
data = data.values
type(data)               # Dataframe to numpy array
X = data[:,0].reshape(-1,1)
Y = data[:,1]
print("X-Shape: ", X.shape)
print("Y-Shape: ", Y.shape)
N,D = X.shape
```

```python
# Scatter plot
plt.scatter(X,Y)
plt.xlabel(' Years')
plt.ylabel('Transistors')
```

```python
# We can see that it's exponential growth, so to make it linear, we take log both sides, in the equation C=A.r^t 
#So it becomes log C= t * log r + log A
# t is the input here, and log A is the constant, and it becomes a linear function
# so we plot t v/s log C, which here is, log Y

Y = np.log(Y)
plt.scatter(X,Y)
plt.xlabel(' Years')
plt.ylabel('Transistors')
```

```python
# Pre- Process values of X, as they are too large
X= X - X.mean()
print(X,Y)
```

```python
# Build the model

model = tf.keras.models.Sequential([
    tf.keras.layers.Input( shape = (D,)),
    tf.keras.layers.Dense(1)
])

#compile the model, use stochaistic gradient descent and 'mse', Mean Squared Error as the loss function

model.compile( optimizer = tf.keras.optimizers.SGD(0.001,0.9), loss='mse')

# Learning rate scheduler

def schedule(epochs, lr):
    # lr is the current learning rate
    if epochs <= 50:
        return 0.0001
    else:
        return 0.001
    
scheduler  = tf.keras.callbacks.LearningRateScheduler(schedule) # scheduler defined
```

```python
#Train the model

r = model.fit(X,Y,epochs=200 ,callbacks=[scheduler])
```

```python
# Plot loss per iteration:
plt.plot(r.history['loss'], label='loss')
```

```python
#Get the slope
a = model.layers[0].get_weights()[0][0,0]
c = model.layers[0].get_weights()[1][0]
print(a,c)
```

```python
# Plot the data and line
plt.scatter(X,Y)
plt.plot(X, a*X + c,color = 'green');
```

```python
# So we find this time required:
print(" The time required to double the count is: ", np.log(2)/a)
```

```python
# Hence, Moore's Law is proved!
```

```python
# Plotting model, using model.predict

Yhat = model.predict(X).flatten()
plt.scatter(X,Y)
plt.plot(X,Yhat, color = 'red');
```
